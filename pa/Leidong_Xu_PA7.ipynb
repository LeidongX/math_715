A program to perturb matrices and plot their eigenvalues 

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from numpy import random as rdm
from numpy import linalg as LA
```

```python
def eigscat(a,err,m):
    
    """
    
    The function `eigscat` plots eigenvalues of real and complex perturbations of a given matrix.
    
    Input: 
        a - square matrix
        
        err - list of errors values
        
        m - number of perturbations
        
    Output: 
        
        - print eigenvalues of the perturbed matrix and their condition numbers
        
        - plot three diagrams of real, complex, and combined perturbations (for every value of the error).
    
    """
    
    import numpy as np
    
    #record the size of the input matrix
    n = a.shape[0] 
    
    #compute the eigenvalues and eigenvectors of a 
    w,v = LA.eig(a)
    
    #normalixe the eigenvectors
    for i in range(n):
        v[i]=v[i]/LA.norm(v[i])
        
    #invert the matrix of normalized eigenvectors
    vi=LA.inv(v) 
    
    #compute and print the condition number of each eigenvalue
    cnd = np.zeros(n)
    
    for i in range(n):
        cnd[i] = LA.norm(vi[i]) 
                
    print ('       Eigenvalues')
    print (w)
    print ('       Condition Numbers') 
    print (cnd)
    print
    
    #compute perturbed eigenvalues
    for error in err:
        
        #print current error
        print ('Error =', error)
        print
        
        #initialize real perturbation er and complex perturbation ei
        er=w 
        ei=w

        for i in range(m):
            #sample real uniform random error matrix
            r=rdm.rand(a.shape[0],a.shape[1])-.5 

            #normalize error matrix to have norm err
            r=(float(error)/LA.norm(r,2))*r

            #compute perturbed eigenvalues
            er = np.append(er,LA.eig(a+r)[0])

        for i in range(m):
            #sample complex uniform random error matrix
            r=(rdm.rand(a.shape[0],a.shape[1])-.5)+1j*(rdm.rand(a.shape[0],a.shape[1])-.5)

            #normalize error matrix to have norm err
            r=(float(error)/LA.norm(r,2))*r

            #compute perturbed eigenvalues
            ei= np.append(ei,LA.eig(a+r)[0])


        #Plot data
        plt.figure(figsize=(30,10))

        #real perturbations
        plt.subplot(1,3,1)
        plt.plot(er.real,er.imag,'xr')
        plt.plot(w.real,w.imag,'ok', linewidth = 3)
        plt.title('Real Perturbations', fontsize=30)

        #complex perturbations
        plt.subplot(1,3,2)
        plt.plot(ei.real,ei.imag,'.b')
        plt.plot(w.real,w.imag,'ok', linewidth = 3)
        plt.title('Complex Perturbations', fontsize=30)

        #real and complex perturbations
        plt.subplot(1,3,3)
        plt.plot(ei.real,ei.imag,'.b')
        plt.plot(er.real,er.imag,'xr')
        plt.plot(w.real,w.imag,'ok', linewidth = 3)
        plt.title('Real and Complex Perturbations', fontsize=30)
        plt.show()
```

#Using newton-schulz iteration to build a precondition $P^{-1}$ then see what happened on condition number of $P^{-1}A$. $P^{-1}$ would be better if it is closed to $A^{-1}$
#and compare with $A^{-1}A$ at last


```python
m =  100
a = np.diag([1,2,2,3,3,3])+np.diag([0,1,0,1e2,1e2],k=1)+np.diag([0,0,0,1e4], k=2)
err = [1e-10, 1e-8, 1e-6, 1e-4, 1e-3]


V= np.diag(1.0/a.diagonal())
def NS(A,V):
    return V.dot(2*np.ones_like(A)-A.dot(V))
PA = NS(a,V)
eigscat(PA.dot(a),err,m)
```
