```python
# import python libraries

%matplotlib inline

import numpy as np
from numpy import random as rdm
from numpy import linalg as LA
import matplotlib.pyplot as plt
```

```python
# number of tests to perform
ntest = 1000

# vectors for storing results
condi = np.zeros(ntest)
error = np.zeros(ntest)
N = 30

for test in range(ntest):
    # random value for delta
    S = np.zeros([N,N])
    for i in range(N):
        S[:,i] = rdm.normal(size=N)
        
    # random number list in [1,2]
    Diag = rdm.rand(N) + 1
    
    u = rdm.uniform(0,1)
    # change the smallest elements at the eigenvalue array
    Diag[Diag==np.amin(Diag)] *= 10**(-u*16)
    # make a diagnoal matrix
    D = np.diag(Diag) 
    
    # create A and solution vector
    A = S.dot(D).dot(np.linalg.inv(S))
    x = rdm.rand(N)

    # compute the right-hand side
    b = np.dot(A,x)

    # solve
    z = LA.solve(A,b)

    # compare
    condi[test] = LA.cond(A)
    error[test] = LA.norm(x-z)/LA.norm(x)
    
```

```python
# get machine epsilon
eps = np.finfo(float).eps

plt.figure(figsize=(10,5))


# plot condi against error
plt.subplot(1,2,2)
t = np.array([ 1, 1e17])
plt.loglog( condi, error, '.', t, eps*t, 'r')
plt.xlabel( 'condition number' )
plt.ylabel( 'rel err' )
```
