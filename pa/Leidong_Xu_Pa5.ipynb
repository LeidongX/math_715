```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import normal
from numpy import linalg as LA
```

Part(A) Lets try one sample first

```python
N = 8
M = 1
plt.figure(figsize=(10,5*N/2.0))
for j in range(N):
    plt.subplot(N/2.0,2,j+1)
    m = 2**(j+1)
    for k in range(M):
        A = normal(size=(m,m))/np.sqrt(m)
        v,w = LA.eig(A)
        plt.scatter(w.real,w.imag)    
```

```python
N = 8
M = 100
plt.figure(figsize=(10,5*N/2.0))
for j in range(N):
    plt.subplot(N/2.0,2,j+1)
    m = 2**(j+1)
    for k in range(M):
        A = normal(size=(m,m))/np.sqrt(m)
        v,w = LA.eig(A)
        plt.scatter(w.real,w.imag)    
```

Part(B)

```python
N = 6
M = 100
#plt.figure(figsize=(10,5*N/2.0))

sr_list = []
#norm_list = []
singular_max_list = []
singular_min_list = []
#cond_list =[]

spectral = 0
singular_max = 0
singular_min = 0
for j in range(N):
    #plt.subplot(N/2.0,2,j+1)
    m = 2**(j+1)
    for k in range(M):
        A = normal(size=(m,m))/np.sqrt(m)
        v,w = LA.eig(A)
        spectral += np.amax(abs(v))
        singular_max += LA.norm(A,ord = 2)
        singular_min += LA.norm(A,ord = -2)
        plt.figure('1')
        plt.scatter(k,np.amax(abs(v)))
        plt.figure('2')
        plt.scatter(k,LA.norm(A,ord = 2))
        plt.figure('3')
        plt.scatter(k,LA.norm(A,ord = 2))
    sr_list.append(spectral/M)
    singular_max_list.append(singular_max/M)
    singular_min_list.append(singular_min/M)
```

```python
x = [1,2,3,4,5,6]
plt.plot(x,sr_list,x,singular_max_list,x,singular_min_list)
plt.legend(['spectral radius','maxinum singular value','min singular value'])
plt.show()
```

```python
x = [2**1,2**2,2**3,2**4,2**5,2**6]
plt.plot(x,sr_list,x,singular_max_list,x,singular_min_list)
plt.legend(['spectral radius','maxinum singular value','min singular value'])
plt.show()
```

```python
N = 8
M = 100
x = [32**-1,16**-1,8**-1,4**-1,2**-1,]
spectral = 0
singular_max = 0
singular_min = 0
for j in range(N):
    #plt.subplot(N/2.0,2,j+1)
    m = 2**(j+1)
    singular_min_list = []
    for k in range(M):
        A = normal(size=(m,m))/np.sqrt(m)
        singular_min += LA.norm(A,ord = -2)
        singular_min_list.append(LA.norm(A,ord = -2))

    fig, ax = plt.subplots()
    plt.hist(singular_min_list,bins= x)
    plt.xticks(x)
    ax.set_xscale('log', basex=2)
    plt.title('m = 2^'+str(j+1))
    plt.show()
```
