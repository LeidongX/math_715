# Gauss method for updating geodetic databases

```python
# Useful libraries

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from numpy import linalg as LA
```

## Reading the database of locations

First we need to read the file `z_db.dat` which contains the landmark locations in the original database. We will use the built-in function [`open()`](https://docs.python.org/2/library/functions.html#open) and we will create a list `z_db` containing a large number of planar points (with an $x$ and a $y$ coordinate). We will let $N$ be the total number of points.

```python
# name of the file to open
fname = 'z_db.dat'

N = 500

z_db = []

with open(fname,'r') as f:

    line_number = 0
    
    # looping over lines in a file
    for line in f:
      
        if line_number == 0:
            print(line)
            
        else:
            print (line_number, line)   ## first time you run it, uncomment this line and comment the line below.
            z_db.append( [float(x) for x in line.split()] )
            
        line_number += 1

z_db
```

## Reading the database of measurements

Here we will read the file `lengths.dat` and use it to create a [dictionary](https://docs.python.org/2/tutorial/datastructures.html#dictionaries), we call `Mmt`, where the keys are pairs of indeces, corresponding to the list of points `z_db` created above, thus corresponding to pairs of points in the plane; and the values of these keys are the distances between those two points. We will let $M$ be the total number of length measurements. Notice our use of the built-in function [`zip()`](https://docs.python.org/2/library/functions.html#zip).

```python
# name of the file to open
fname = 'lengths.dat'

M = 2333

ind = []
L_ij = []


with open(fname,'r') as f:

    line_number = 0
    
    # looping over lines in a file
    for line in f:
        if line_number == 0:
            print(line)
            
        else:
            print (line_number, line)   ## first time you run it uncomment this line and comment the line below.
            b = line.split()
        
            ind.append( tuple( int(y)-1 for y in b[0:-1] ) )
            L_ij.append( float(b[-1]) )
            
        line_number += 1

Mmt = dict(zip(ind,L_ij))

Mmt
```

Here is a nice way to create, from the dictionary of measurements, the list of pairs whose distance was measured.

```python
ind = [keys for keys, values in Mmt.items()]

ind
```

And here is a way to create, from the dictionary of measurements, the list of distances that were measured.

```python
L = [values for keys, values in Mmt.items()]

L
```

## Reading the database of measurements "true" positions

Finally, we read the file `z_true.dat` containing the data with the the true positions, which we store in `z_true`.

WARNING!! This is for comparison only.  DO NOT USE THIS ARRAY in your calculations.

```python
# name of the file to open
fname = 'z_true.dat'

N = 500

z_true = []

with open(fname,'r') as f:

    line_number = 0
    
    # looping over lines in a file
    for line in f:
      
        if line_number == 0:
            print(line)
            
        else:
            #print line_number, line   ## first time you run it uncomment this line and comment the line below.
            z_true.append( [float(x) for x in line.split()] )
            
        line_number += 1

z_true
```

## Computing the original lengths

We first make a copy of the original positions and then compute the lengths using the original positions in the database. Note that since `z_db` is a list of lists, given a pair of indeces `(u,v)` the expression `z_orig[u]-z_orig[v]` does not make sense because lists cannot be subtracted. So we first have to transform them into `np.array` objects.

```python
# make a copy of original positions
z_orig = z_db

# compute the lenghts in the original database corresponding to the true measurements 
L_orig = []

for (u,v) in ind:
    L_orig.append( LA.norm( np.array(z_orig[u])-np.array(z_orig[v]), 2) )
    #print (u,v), LA.norm( np.array(z_orig[u])-np.array(z_orig[v]), 2)

L_orig
```

## Comparing the true positions to the original ones

We now compare the true positions to the database positions. Notice how the first two are normailzed to be the same.

```python
np.array(z_true) - np.array(z_db)
```

In order to compute the magnitude of the error made in the positions we build a vector from all these coordinate errors. The 'list comprehension' way to flatten a list of lists is to write `[item for sublist in list for item in sublist]`

```python
[i for pos in (np.array(z_true) - np.array(z_db)) for i in pos]
```

Now we can simply store the $2$-norm of this vector of coordinate errors.

```python
LA.norm( [i for pos in (np.array(z_true) - np.array(z_db)) for i in pos], 2)
```

To get a relative error we will divide by a similar $2$-norm for the coordinates of the true positions. 

```python
LA.norm( [i for position in z_true for i in position] )
```

## The error in positions

We name the original relative error in positions `Z_err`.

```python
# compute original error norms in the positions
Z_err = LA.norm( [i for pos in (np.array(z_true) - np.array(z_db)) for i in pos], 2) / LA.norm( [i for position in z_true for i in position] )

Z_err
```

## The error in lengths

We already computed the original lengths in `L_orig` from the database of positions. We also have the new measurements stored in `L`.

```python
L_new = L

L_err = LA.norm(np.array(L_new)-np.array(L_orig) ,2) / LA.norm(L_new)

L_err
```

In the next cell, implement a function `improve_g(z, ind, L)`. Its input should consist of a list `z` of plane locations, a list `ind` of tuples of indices corresponding to pairs of positions for which a measurement was surveyed, and a list `L` of such surveyed distances. The function should update the positions `z` using Gauss' method and return new positions `z_updated`. 
</div>
</div>

```python
def improve_g( z, ind, L ):

    import scipy as sp
    from scipy.sparse import csr_matrix
    from scipy.sparse.linalg import lsqr
    
    N = len(z)
    M = len(ind)
    b = np.zeros( (M, 1) )

    row = (4*M)*[0]
    column = (4*M)*[0]
    val = (4*M)*[0]
    

    A = np.zeros( [M, 2*N])
    k = 0
    for (u,v) in ind:
        b[k] = L[k]**2 - LA.norm( np.array(z[u])-np.array(z[v]), 2)**2
        A[k,u*2] = 2*(z[u][0]-z[v][0])
        A[k,u*2 + 1] = 2*(z[u][1]-z[v][1])
        A[k,v*2] = -2*(z[u][0]-z[v][0])
        A[k,v*2 + 1] = -2*(z[u][1]-z[v][1])
        k += 1

    #plt.spy(A)
    A = csr_matrix(A)
    
    x = lsqr(A,b)[0]
    z_new = (np.array(z.flatten()) + x).reshape(-1,2)
    
    return z_new
```

```python
for iteration in range(10):

    # update the approximation
    z_updated = improve_g( z_orig, ind, L )

    # visualize the original, updated, and true locations
    plt.figure(figsize=(10,10))
    
    plt.plot( [z_true[i][0] for i in range(N)], [z_true[i][1] for i in range(N)], 'k*', 
         [z_db[i][0] for i in range(N)], [z_db[i][1] for i in range(N)], 'rx', 
         [z_updated[i][0] for i in range(N)], [z_updated[i][1] for i in range(N)], 'bo'   ) 
        
    plt.title( 'Updated Positions' )
    plt.legend( ('true', 'database', 'updated'), loc='upper right' )

    # compute the updated lengths
    L_updated = []
    for (u,v) in ind:
        L_updated.append( LA.norm( np.array(z_updated[u])-np.array(z_updated[v]), 2) )

    # output errors
    print ('iter:                           {:d}'.format( iteration ))
    
    print ('Original location error norm:   {:e}'.format( Z_err ))
    
    print ('New location error norm:        {:e}'.format( LA.norm( [i for pos in (np.array(z_true) - np.array(z_updated)) for i in pos], 2) / LA.norm( [i for position in z_true for i in position] )
))
    
    print ('Original length error norm:     {:e}'.format( L_err )) 
    
    print ('New length error norm:          {:e}'.format( LA.norm(np.array(L_new)-np.array(L_updated) ,2) / LA.norm(L_new) ))

    print ('\n') 

    z_orig = z_updated
```
